# EDA — Grid Loss Model Dataset
`../data/dataset aggregation/output/agg_model_dataset.csv`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from IPython.display import display

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)

df = pd.read_csv('../data/dataset aggregation/output/agg_model_dataset.csv')
df
```

```python
# filter to selected columns
FEATURES = [
    'flood_ratio',        # flood exposure  — fraction of cell inundated (GeoTIFF)
    'mean_hag',           # vulnerability   — mean first-floor height above ground (ft)
    'pct_elevated',       # vulnerability   — fraction of buildings elevated >1ft
    'mean_age',           # vulnerability   — mean building age (years)
    'res_ratio',          # exposure type   — fraction residential
    'bldg_density',       # exposure        — buildings per km2
    'mean_bldg_value',    # exposure value  — mean replacement value ($)
    'mean_sqft',          # building size   — mean heated sq ft
    'areaSqKm_developed', # land use        — developed area in cell
]

df = df[FEATURES + ['loss_ratio', 'event_name', 'county',
                           'year', 'grid_id', 'total_paid', 'total_bldg_val']].dropna()
```

```python
df_model = df[FEATURES + ['loss_ratio']].dropna()

X = df_model[FEATURES].values
y = df_model['loss_ratio'].values

print(f'Dataset: {X.shape[0]} rows x {X.shape[1]} features')
print(f'Target  : mean={y.mean():.4f}, std={y.std():.4f}, '
      f'range=[{y.min():.4f}, {y.max():.4f}]')
print()
print('Feature descriptions:')
feat_desc = {
    'flood_ratio':        'Flood inundation ratio (GeoTIFF-derived)',
    'mean_hag':           'Mean height above ground = FFE - LIDAR_LAG (ft)',
    'pct_elevated':       'Fraction of buildings with first floor >1ft above ground',
    'mean_age':           'Mean building age (years)',
    'res_ratio':          'Fraction of residential buildings',
    'bldg_density':       'Building density (per km2)',
    'mean_bldg_value':    'Mean building replacement value ($)',
    'mean_sqft':          'Mean heated square footage',
    'areaSqKm_developed': 'Developed land area (km2)'
}
for f, d in feat_desc.items():
    print(f'  {f:<22} -> {d}')
```

```python
df_model.to_csv('../data/final_model_dataset.csv', index=False)
```

## 1. Schema — dtypes, nulls, duplicates

```python
# ── column-level audit ──────────────────────────────────────────────────
schema = pd.DataFrame({
    'dtype':        df_model.dtypes,
    'null_count':   df_model.isnull().sum(),
    'n_unique':     df_model.nunique()
})
display(schema)
```

## 2. Numeric summary stats

```python
num_cols = df_model.select_dtypes(include='number').columns.tolist()

stats = df_model[num_cols].agg(['mean','median','std','min','max']).T
stats['kurtosis'] = df_model[num_cols].kurtosis()
stats['cv']       = (df_model[num_cols].std() / df_model[num_cols].mean()).abs()   # coeff of variation

display(stats.round(4))
```

## 3. Categorical / object columns

```python
cat_cols = df.select_dtypes(exclude='number').columns.tolist()
print('Categorical columns:', cat_cols)

for col in cat_cols:
    print(f'\n── {col} ──')
    vc = df[col].value_counts(dropna=False)
    print(vc.to_string())
```
