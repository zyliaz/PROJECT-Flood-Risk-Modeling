# Data Extraction & Cleaning
**Data Source:** FEMA NFIP Redacted Claims v2 (OpenFEMA), in parquet format.

**Target Area:** 5 eastern NC counties with higher flood risk: Lenoir, Wayne, Robeson, Beaufort, Edgecombe  

```python
import pandas as pd
import duckdb
```

```python
paths = ({
    'filtered_df':'selected_county_claims.csv'
})
```

```python
# extract dataset, filter to 5 target counties
query = """
SELECT *
FROM 'FimaNfipClaimsV2.parquet'
WHERE countyCode in ['37107', '37191', '37155', '37013', '37065']
"""
df = duckdb.query(query).to_df()
```

```python
# export filtered dataset
df.to_csv(paths['filtered_df'], index=False)
```
